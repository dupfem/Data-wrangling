```python
# In order to get the distribution of age and income we need to create bins for each of them

#Age

age_bins = [0,20,30,40,50,90]
age_labels = ['Teens', 'Youth', 'Young Adult','Adult','Elder']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)

# Income

income_bins = [0, 25000, 50000, 75000, 200000]
income_labels = ['low_income', 'medium_low', 'medium_high', 'high_income']
df['Income_Group']= pd.cut(df['Income'], bins=income_bins, labels=income_labels)

```

```python
# Create a new column to indicate the education level
def determine_education(row):
    if row['education_PhD'] == 1:
        return 'PhD'
    elif row['education_Master'] == 1:
        return 'Master'
    elif row['education_Graduation'] == 1:
        return 'Graduation'
    elif row['education_Basic'] == 1:
        return 'Basic'
    else:
        return 'primary'
```

```python
# Code to get family size

col2 = ['marital_Divorced', 'marital_Married','marital_Single', 'marital_Together', 'marital_Widow']
tFSize = []
for sample in df.values:
    numKids = sample[-1]
    
    if sample[26] == 1 or sample[28]==1 or sample[30]==1:
        tFSize.append(numKids+1)
    else: 
        tFSize.append(numKids+2)
```

```python
# code to add colum for type of family
fSize = list(df.fSize)
fSeg = []
for val in fSize:
    if val == 1:
        fSeg.append('mono')
    elif val == 2:
        fSeg.append('duo')
    elif val == 4:
        fSeg.append('Tetra')
    else:
        fSeg.append('mega')

df['f_Seg'] = fSeg
```

```python
# Create a new column to indicate the marital level
def determine_marital(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'    
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    else:
        return 'Widow'
```

```python
df['Marital_Group'] = df.apply(determine_marital, axis=1)
```
